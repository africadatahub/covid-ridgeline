```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
owid = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
df = owid.loc[owid.continent == 'Africa']
```

```python
df = df[['date','iso_code','location','new_cases_smoothed']]
df = pd.pivot(df, index='iso_code', columns=['date'], values=['new_cases_smoothed'])
df.to_csv('new_cases_smoothed.csv')
df
```
